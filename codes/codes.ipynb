```python
import pandas as pd
import numpy as np
```

```python
infos_path = "./data/ingredient_degree.xlsx"
infos = pd.read_excel(infos_path)
```

```python
infos.head(2)
```

```python
columns = infos.columns.to_list()
columns[:5]
```

```python
index = columns[2:36]
index[:5]
```

```python
pericadium_infos = infos[index]
pericadium_infos.head(2)
```

```python
pericadium_infos["Na-Na+"] = pericadium_infos["Na"]-pericadium_infos["Na+"]
pericadium_infos["K-K+"] = pericadium_infos["K"]-pericadium_infos["K+"]
pericadium_infos["Ca-Ca2+"] = pericadium_infos["Ca"]-pericadium_infos["Ca2+"]
pericadium_infos["Mg-Mg2+"] = pericadium_infos["Mg"]-pericadium_infos["Mg2+"]
pericadium_infos.head(2)
```

```python
infos["label_num"] = [num.split("（")[0] for num in infos["循环缺失"].values]

infos["label_num"]= infos["label_num"].str.strip("%").astype(float)/100

```

```python
pericadium_infos.head(2)
```

```python
category_ls = []
for num in infos["label_num"]:
    if num == 0:
        category_ls.append(0)
    elif num >0 and num <= 0.333:
        category_ls.append(1)
    elif num > 0.33 and num <=0.666:
        category_ls.append(2)
    else:
        category_ls.append(3)
pericadium_infos["label"] = category_ls
        
```

```python
pericadium_infos.head(2)
```

```python
pericadium_infos.to_excel("./data/sub_data/cycle_defeat_infos.xlsx")
```

```python
##计算每一个特征与目标向量之间的方差分析F值
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import f_classif
```
